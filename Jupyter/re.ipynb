## 正则表达式对字符串的逻辑操作，主要是对字符串的一种过滤，

## 用“元字符” 与“普通字符”组成一个字符串规则对已知的字符串或文本过滤出自己想要的字符串


# 元字符：





# 常用函数：match  search findall sub split 


# 1: re.match(pattern, string, [flags]) 

# 从字符串开始处进行匹配,在起始位置匹配成功，返回Match对象，如果不符合，匹配失败，返回None


## pattern:匹配规则  string:要匹配的字符串 flags:参数


```python
import re
pattern = r'(13[4-9]\d{8})$|(15[01289]\d{8})$'
num = input('手机号码： ')
match = re.match(pattern,num)
if match == None:
    print('不是移动号码')
else:
    print('是移动号码')
```

# 2:re.search(pattern,string,[flags])
# 在整个字符串中搜索第一个匹配的值，成功，返回Match对象,否则返回None



```python
import re
pattern = r'(习)|(近)|(平)'
string = input('信息：')
match = re.search(pattern,string)
if match == None:
    print(string,'可以发布')
else:
    print('出现敏感词汇！')
```

#  3:re.findall(pattern,string,[flags])

# 返回的结果是一个列表，建中存放的是符合规则的字符串，如果没有符合规则的字符串被找到，就会返回一个空值

```python
re.findall(r"hello", "hello world hello")  # 普通匹配
```

```python
re.findall(r"^hello", "hello world hello")  # 匹配开头
```

```python
re.findall(r"^hello", "hello world\nhello")  # 多行匹配开头
```

```python
re.findall(r"hello$", "hello world hello")  # 匹配结尾
```

```python
re.findall(r"\d", "aaa111bbb222ccc333")  # 匹配数字
```

```python
re.findall(r"\d{2}", "aaa111bbb222ccc333")  # 匹配两位的数字
```

```python
re.findall(r"ab|cd", "ab000cd00")  # 匹配"ab"或"cd"
```

```python
re.findall(r"(abc)+", "abcabcabc")  # 想要匹配多个"abc"，使用分组时会优先把分组的内容返回
```

```python
re.findall(r"(?:abc)+", "abcabcabc")  # 想要匹配多个"abc"，(?:)把分组去掉，变成一个普通的字符串
```

# 4:sub(pattern,repl,string,[count],[flags])

# 字符串替换

```python
pattern = r'(习)|(近)|(平)'
string = input('发言：')
sub = re.sub(pattern,'*',string)
print(sub)
```

# 5:split(pattern,string,[maxsplit],[flags])

# 根据正则表达式分割字符串。并以列表形式返回

```python
str1 =input('好友（@**）')
pattern = r'@'
list1 = re.split(pattern,str1)
print('您将通知以下好友：')
for i in list1:
    print(i)
```

# 贪婪匹配  非贪婪匹配

## 贪婪匹配总是匹配到最长的那个字符串，相应的，非贪婪匹配是匹配到最小的那个字符串，只需要在匹配字符串的时候加一个？即可

```python
string = 'Beautiful is better than ugly.Explicit is better than impliciy.'
obj1 = re.compile('Beautiful.*y\.')
print(obj1.findall(string))
obj2 = re.compile('Beautiful.*?\.')
print(obj2.findall(string))
```

# 小括号分组

## 如果是要返回括号中匹配的字符串中，则该小括号不需要转义符，findall方法返回的是小伙好中匹配到的字符串，search.group（）方法的返回的整个模式匹配到字符串，search.group(1)这个是匹配第一个小括号中的模式匹配到的字符串，search.group(2)这个是匹配第二个小括号中的模式匹配到的字符串，以此类推


```python
string = '=aa12345d&& 0a () --'
obj = re.compile('(a).*(&)')
print(obj.findall(string))
print(obj.search(string).group())
print(obj.search(string).group(1))
print(obj.search(string).group(2))

```
